4.1 Apply and functions

Let's say we want to calculate the deviation of mean review points. Firsty, let's take a look at the data in depth

```python
wine['points']
```

```python
wine['points'].describe()
```

we could get the mean of all scores and calcute the deviation

```python
wine['points'].mean()
```

```python
# show the calculation
wine['points'] - wine['points'].mean()
```

```python
# store it as a new column
wine['std_points'] = wine['points'] - wine['points'].mean()
```

```python
# verify if it exists
wine['std_points'].describe()
```

What if we cannot create a new series? In this case, we will convert review points from a hundred into a ten-point scale. <br>
We could consider using function.

```python
def standardize(pts):
    return pts/10
```

Next to that, we will apply this function to the dataset

```python
wine['std_points'] = wine['points'].apply(standardize) 
wine['std_points'].head()
```

Don't want to write a long code? Let's use lambda function

```python
wine['std_points'] = wine['points'].apply(lambda pts: pts/10)
wine['std_points'].head()
```

With this idea of using apply() and functions, we could also do pretty much of things. See this example of defining interpretation of points scoring.

- If Score >= 95: very excellent
- If Score >= 90: excellent
- If Score >= 85: good
- Othewise: acceptable

```python
# create a function to define way to interpret scores
def interpret_score(pts):
    if pts >= 95:
        return 'very excellent'
    elif pts >= 90:
        return 'excellent'
    elif pts >= 85:
        return 'good'
    else:
        return 'acceptable'
```

```python
# apply on the dataset
wine['interpretation'] = wine['points'].apply(interpret_score) 
wine[['points', 'interpretation']].tail(10)
```
